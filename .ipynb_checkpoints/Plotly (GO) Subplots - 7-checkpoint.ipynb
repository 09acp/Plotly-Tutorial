```python
from plotly.offline import iplot, init_notebook_mode
#from plotly.offline import plot # FOR SPYDER
import plotly.graph_objects as go
from plotly import tools
import plotly 

import random
import numpy as np
np.random.seed(122)
import pandas as pd
```

```python
female_age = np.random.randint(low = 21, high = 75, size = 100) #lowest,highest
male_age = np.random.randint(low = 3, high = 100, size = 100) #lowest,highest

female_salary = np.random.uniform(low = -10, high = 10, size = 100) #lowest,highest
male_salary = np.random.uniform(low = -10, high = 10, size = 100) #lowest,highest

female_debt = male_debt = np.random.randint(15,30,100)
female_tax = male_tax = np.random.randint(100) + 10
```

## Examples

```python
# MAKING 2 SUBPLOTS ON SAME ROW
def make_trace( X, Y, colour, Name):
    return go.Scatter(x = X,
                      y = Y,
                      mode = "markers",
                      name = Name,
                      marker = dict(size =12,
                                    color = colour)  )

trace0 = make_trace(female_salary, female_age, "rgba(255,0,70,1)" , "Female")    
trace1 = make_trace(male_salary, male_age, "rgba(0,190,70,1)" , "Male")

" DEPRECIATED "
# fig = tools.make_subplots(rows = 1,
#                           cols = 2,
#                           subplot_titles = ("Female","Male") )                         

fig = plotly.subplots.make_subplots(  rows = 1,
                                      cols = 2,
                                      subplot_titles = ("Female","Male") )  


fig.append_trace( trace0, 1, 1) # row 1, col 1
fig.append_trace( trace1, 1, 2)

fig.layout.update( title = "Correlation between A & B ")
iplot(fig);
```

```python
# SHARING AN AXIS                    
fig = plotly.subplots.make_subplots(rows = 1,
                                    cols = 2,
                                    subplot_titles = ("Female","Male"),
                                    shared_yaxes = True,)  # CAN BE x
                  
fig.append_trace( trace0, 1, 1) # row 1, col 1
fig.append_trace( trace1, 1, 2)

fig.layout.update( title = "Correlation between A & B ",
                   yaxis = dict(title = "Age" )  )
iplot(fig);
```

```python
# SHARING X AXIS 
fig = plotly.subplots.make_subplots(rows = 2,
                                    cols = 1,
                                    subplot_titles = ("Female","Male"),
                                    shared_xaxes = True,)  # CAN BE x
                  
fig.append_trace( trace0, 1, 1) # row 1, col 1
fig.append_trace( trace1, 2, 1)

fig.layout.update( title = "Correlation between A & B ",
                   xaxis = dict(title = "Salary" )  )
iplot(fig);
```

```python
# CONSTRAIN PLOT PROPORTIONS
trace0 = go.Scatter(x= female_salary,
                    y = female_salary,
                    mode = "markers",
                    name = "Female",
                    marker = dict(size = 12,

                                  color = "rgba(244,0,0,1)")  )
trace1 = go.Scatter(x= male_salary,
                    y = male_salary,
                    mode = "markers",
                    name = "Male",
                    marker = dict(size = 12,
                                  color = "rgba(0,200,0,1)"),
                    xaxis = "x2",
                    yaxis = "y2",)

data = [ trace0, trace1]

# 70% x-axis plot 1, 20% x-axis plot 2
layout = go.Layout(title = "CONSTRAIN X AXIS PROP",
                   xaxis = dict(domain = [0,0.7]), # 0-1  70% of left side
                   xaxis2 = dict(domain = [0.8 , 1]), # 20% on right side
                   yaxis2 = dict(anchor = "x2")        )

fig = go.Figure(data = data,
                layout = layout,  )

iplot(fig)
```

```python
# CUSTOMIZE AXIS 
trace0 = go.Histogram(x=female_age)
trace1 = go.Histogram(x=female_age)
trace2 = go.Histogram(x=female_age)
trace3 = go.Histogram(x=female_age)

fig = plotly.subplots.make_subplots(rows=2,
                                    cols=2,
                                    subplot_titles=("FigA",
                                                    "FigB",
                                                    "FigC",
                                                    "FigD",)    )
fig.append_trace(trace0,1,1)
fig.append_trace(trace1,1,2)
fig.append_trace(trace2,2,1)
fig.append_trace(trace3,2,2)

fig.layout["xaxis1"].update(title = "Female age")
fig.layout["xaxis2"].update(title = "Male age")
fig.layout["xaxis3"].update(title = "Female sal")
fig.layout["xaxis4"].update(title = "Male sal")

fig.layout["yaxis1"].update(title = "Frequency")
fig.layout["yaxis2"].update(title = "Frequency")
fig.layout["yaxis3"].update(title = "Frequency")
fig.layout["yaxis4"].update(title = "Frequency")

fig.layout.update(title="Hists age/sal according to age",
                  showlegend=False  )
iplot(fig)
```

```python
# ODD PAIRING
## 2 small plots
### 1 large plot
trace0 = go.Histogram(x=female_age)
trace1 = go.Histogram(x=male_age)
trace2 = go.Histogram(x=male_salary)

# col 2 spans entire column
fig = plotly.subplots.make_subplots(rows=2,
                                    cols=2,
                                    specs= [ [ {}, {}], [{"colspan":2}, None]],
                                    subplot_titles=("FigA",
                                                    "FigB",
                                                    "FigC",)    )
fig.append_trace(trace0,1,1)
fig.append_trace(trace1,1,2)
fig.append_trace(trace2,2,1)

fig.layout["xaxis1"].update(title = "Female age")
fig.layout["xaxis2"].update(title = "Male age")
fig.layout["xaxis3"].update(title = "Debt")

fig.layout["yaxis1"].update(title = "Frequency")
fig.layout["yaxis2"].update(title = "Frequency")
fig.layout["yaxis3"].update(title = "Frequency")

fig.layout.update(title="Hists age/sal according to age",
                  showlegend=False  )
iplot(fig)
```
